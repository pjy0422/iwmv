```python
import json
import unicodedata


def normalize_text(text):
    # Normalize unicode characters
    normalized_text = unicodedata.normalize("NFKD", text)
    # Remove special characters like soft hyphen
    cleaned_text = "".join(
        c
        for c in normalized_text
        if not unicodedata.category(c).startswith("Cf")
    )
    return cleaned_text.strip()


# Load the original dataset
with open("data/contriever_nq_test.json", encoding="utf-8") as f:
    data = json.load(f)

# Extract and normalize question-answer pairs
new_data = []
for item in data:
    question = normalize_text(item["question"])
    answer = normalize_text(item["answers"][0])
    new_pair = {"question": question, "answer": answer}
    new_data.append(new_pair)

# Save the cleaned dataset
with open("data/qa_only.json", "w", encoding="utf-8") as f:
    json.dump(new_data, f, indent=2, ensure_ascii=False)

# Verify the saved data
with open("data/qa_only.json", "r", encoding="utf-8") as f:
    data = json.load(f)
    print(data[:10])  # Display the first 10 items to verify
```
