```python
original_data_path = "/home/guest-pjy/data/0829/0828_NQ_filtered.json"
new_data_path = "/home/guest-pjy/data/0901/0901_NQ_poisonedrag.json"
import os
from utils.json_utils import *

os.makedirs("/home/guest-pjy/data/0901", exist_ok=True)
data = load_json(original_data_path)
```

```python
for item in data:
    counterfactual = item["counterfactual"]
    question = item["question"]
    if question.endswith("?"):
        question = question + " "
    elif question.endswith("."):
        question = question + " "
    elif question.endswith("!"):
        question = question + " "
    elif question.endswith(" "):
        question = question + ", "
    else:
        question = question + ", "

    for cf in counterfactual:
        cf["contexts"] = [question + ctx for ctx in cf["contexts"]]
    item["counterfactual"] = counterfactual
```
